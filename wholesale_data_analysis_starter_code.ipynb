## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
def subtotal(row):
    return row['unit_price'] * row['qty']
df['Subtotal'] = df.apply(subtotal, axis=1)
```

```python
df[['unit_price','Subtotal']].head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
def shipping(row):
    if row['unit_weight'] * row['qty'] > 50:
        return 7 * row['unit_weight'] * row['qty']
    return 10 * row['unit_weight'] * row['qty']

df['shipping'] = df.apply(shipping, axis=1)
```

```python
df[['unit_price','Subtotal','shipping']].head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
def total_price(row):
    total_price = row['Subtotal'] + row['shipping']
    total_price += total_price * (9.25/100)
    return total_price

df['total_price'] = df.apply(total_price, axis=1)
```

```python
df[['unit_price','Subtotal','shipping','total_price']].head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).

def s_subtotal(row):
    return (row['unit_cost'] * row['qty']) + row['shipping']

df['unit_total'] = df.apply(s_subtotal, axis=1)
```

```python
df[['unit_price','Subtotal','shipping','total_price','unit_total']].head()
```

```python
# Create a column for the profit of each line using line cost and line price
df['profit'] = df['total_price'] - df['unit_total']
```

```python
df[['unit_price','Subtotal','shipping','total_price','unit_total','profit']].head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.

## QUESTION AMBIGUOUS, USING CLIENT ID'S FROM PART 1

for i in range(0,5):
    print(f"{i}. Client {client_ordered_count[0][i]} spent a total of ${df.loc[df['client_id'] == client_ordered_count[0][i], 'total_price'].sum():,.2f}\n")
    
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit.

top_client_list=[]

for i in range(0,5):
    client_dict = {
        'client_id' : client_ordered_count[0][i],
        'total units purchased' : df[df['client_id'] == client_ordered_count[0][i]]['qty'].sum(),
        'total shipping price' : df[df['client_id'] == client_ordered_count[0][i]]['shipping'].sum(),
        'total revenue' : df[df['client_id'] == client_ordered_count[0][i]]['Subtotal'].sum(),
        'total profit' : df[df['client_id'] == client_ordered_count[0][i]]['profit'].sum()
    }
    top_client_list.append(client_dict)
```

```python
top_client_list_df = pd.DataFrame(top_client_list)
```

```python
top_client_list_df
```

```python
# Format the data and rename the columns to names suitable for presentation.
top_client_list_df.rename(columns = {"client_id":"Client ID",
                                     "total units purchased":"Units",
                                     "total shipping price":"Shipping",
                                     "total revenue":"Revenue",
                                     "total profit":"Profit"}, inplace=True)
```

```python
top_client_list_df
```

```python
# Define the money columns. 
money_columns = ['Shipping','Revenue','Profit']
```

```python
# Define a function that converts a dollar amount to millions.
def currency_format_millions(value):
    return round(value/1000000,1)
```

```python
# Apply the currency_format_millions function to only the money columns. 
for col in money_columns:
    top_client_list_df[col] = top_client_list_df[col].apply(currency_format_millions)
```

```python
# Rename the columns to reflect the change in the money format. 

top_client_list_df.rename(columns = {"Shipping":"($M)Shipping","Revenue":"($M)Revenue","Profit":"($M)Profit"}, inplace=True)
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
top_client_list_df.sort_values(by ="($M)Profit", ascending=False, inplace=True)
```
